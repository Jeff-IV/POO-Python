Importamos las librerias necesarias para el ejercicio


```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.messages import HumanMessage, SystemMessage
```

Creamos una clase abstracta que va a servir como interfaz para no depender de una herramienta especifica en la clase principal (AnalizadorSentimiento)

```python
from abc import ABC
class Analizador(ABC):
  @classmethod
  def analizar(self, texto):
    pass
```

Creamos la clase con la herramienta que usaremos para implementar la clase base


```python
class ModeloGenAI(Analizador):
  """
  Representa un ModeloGENAI que puede analizar sentimientos de texto.
  Hereda de la clase abstracta Analizador.

  Attributes:
    llm (GoogleGenerativeAI): representa el modelo de lenguaje grande que hara de experto analizador.
    contexto (SystemMessage): representa el contexto del analizador.
  """
  def __init__(self):
    self.llm = llm = GoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0.6
    )
    self.contexto = SystemMessage("Eres un analizador de sentimientos. Basandote en el texto recibido responde que sentimientos se expresan.")

  def analizar(self, texto):
    """
    Analiza los sentimientos expresados en un texto.

    Args:
      texto (str): El texto a analizar.

    Returns:
       Un string con los sentimientos expresados en el texto.
    """
    mensajes = [self.contexto, HumanMessage(texto)]
    respuesta = self.llm.invoke(mensajes)
    return respuesta
```

Creamos la clase principal que se encargara de utilizar las herramientas para analizar los sentimientos

```python
class AnalizadorSentimiento():
  """
  Representa un AnalizadorSentimiento que puede analizar sentimientos de texto.

  Attributes:
    herramienta (Analizador): representa la herramienta que hara de experto analizador.
  """
  def __init__(self):
    self.herramienta = ModeloGenAI()

  def analisis(self, texto):

    """
    Usa la herramienta para analizar los sentimientos en un texto.

    Args:
      texto (str): El texto a analizar.

    Returns:
       Un string con los sentimientos expresados en el texto.
    """
    return self.herramienta.analizar(texto)
```

Flujo principal con el programa para el usuario


```python
analizador = AnalizadorSentimiento()
print("-"*15 + " ANALIZADOR DE SENTIMIENTOS " + "-"*15)
analisis = analizador.analisis(input("Escribe como ha sido tu dia: "))
print(analisis)
```
